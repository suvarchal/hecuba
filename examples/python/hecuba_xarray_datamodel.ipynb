```python
import numpy as np
import xarray as xr
import models
```

## Load a sample netcdf dataset from a file

A netCDF may contain many dimensioned variables, this dataset has just one data variable air and time, lat, lon dimensions.

```python
da = xr.open_dataset('test_dataset.nc')
```

```python
da
```

## Save a single variable of netCDF4 or dataarray in xarray parlance

```python
hecuba_tarr = models.DataArray.from_xr_dataarray(da.air) # make a xrdataarray into hecuba compatable dataarray
```

```python
hecuba_tarr.make_persistent('tair') # make persistent
```

```python
models.DataArray.to_xr_dataarray('tair2') # retrive data from cassandra using hecuba
```

## Save an entire netCDF dataset
currently just one variable, air but can be many variables.

```python
hecuba_air_da = models.Dataset.from_xr_dataset(da) # returns hecuba dataset obj
```

```python
hecuba_air_da.make_persistent('air_dataset') # make persistent
```

```python
models.Dataset.to_xr_dataset("air_dataset") # retrieve data from cassandra with hecuba
```
